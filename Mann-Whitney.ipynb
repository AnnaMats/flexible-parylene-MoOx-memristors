```python
import sys
from scipy.stats import mannwhitneyu
from pathlib import Path
```

```python
COLUMN_NAMES = ["initial", "2 cm", "1.5 cm", "1 cm", "0.5 cm"]
```

```python
def read_columns(path, missing="--"):
    columns = {name: [] for name in COLUMN_NAMES}
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            if not line.strip():
                continue
            parts = line.split()          
            for name, cell in zip(COLUMN_NAMES, parts):
                if cell != missing:      
                    columns[name].append(float(cell))
    return columns
```

```python
def rank_biserial(u_statistic, n1, n2):
    return 1.0 - (2.0 * u_statistic) / (n1 * n2)
```

```python
def calc_mann_whitneyu_rb(columns):
    for i in range(1, len(COLUMN_NAMES) - 1):
        name1, name2 = COLUMN_NAMES[i], COLUMN_NAMES[i + 1]
        x, y = columns[name1], columns[name2]

        u_stat, p_value = mannwhitneyu(x, y, alternative="two-sided")
        rb = rank_biserial(u_stat, len(x), len(y))

        print(f"{name1} vs {name2}")
        print(f"U-statistic = {u_stat:.1f}")
        print(f"p-value = {p_value:.6g}")
        print(f"rank-biserial r = {rb:.4f}")
        print("-" * 30)

    print("-" * 30)

    for i in range(len(COLUMN_NAMES) - 1):
        name1, name2 = COLUMN_NAMES[0], COLUMN_NAMES[i + 1]
        x, y = columns[name1], columns[name2]

        u_stat, p_value = mannwhitneyu(x, y, alternative="two-sided")
        rb = rank_biserial(u_stat, len(x), len(y))

        print(f"{name1} vs {name2}")
        print(f"U-statistic = {u_stat:.1f}")
        print(f"p-value = {p_value:.6g}")
        print(f"rank-biserial r = {rb:.4f}")
        print("-" * 30)
```

```python
columns = read_columns(Path("Uset.txt"))
print("U_set")
    
print("Sizes of the distributions:")
for name in COLUMN_NAMES:
    print(f"{name}: {len(columns[name])}")
print()
calc_mann_whitneyu_rb(columns)
```

```python
columns = read_columns(Path("Ureset.txt"))
print("U_reset")
    
print("Sizes of the distributions:")
for name in COLUMN_NAMES:
    print(f"{name}: {len(columns[name])}")
print()
calc_mann_whitneyu_rb(columns)
```
